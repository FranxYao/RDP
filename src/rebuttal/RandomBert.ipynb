```python
random_bert_log = '/home/s1946695/RDP/outputs/bertnet_0.3.0.0/train_log.txt'
normal_bert_log = '/home/s1946695/RDP/local/bertnet_0.0.4.10.4_outputs/train_log.txt'
```

```python
# random bert traing p(x | z)

def read_p_x(filename):
    log_p_x = []
    with open(filename) as fd:
        lines = fd.readlines()
        for l in lines: 
            if(l.startswith('loss')):
                l = l.split(', ')
                p_x = float(l[3].split(': ')[1])
                log_p_x.append(p_x)
    return log_p_x
```

```python
random_bert_p_x = read_p_x(random_bert_log)
normal_bert_p_x = read_p_x(normal_bert_log)
```

```python
plt.plot(random_bert_p_x[5:300], label='Random encoder')
plt.plot(normal_bert_p_x[5:300], label='Bert encoder')
plt.ylabel('p(x | z)')
plt.legend()
```
